```python
""" Projeto Lógica de Programação - Phyton 
    Introdução a Programação - Thiago Pires
             Diário Professor               """

# Dicioário para armazenar alunos e suas notas
alunos = {}

# Função adiciona novo aluno ao diário do professor
def novo_aluno(alunos, ra, nome, notas):
    if ra in alunos:
        print(f"\nErro: O RA {ra} já está em uso. Por favor, insira um RA diferente.")
    else:
        alunos[ra] = {'nome': nome,'notas': notas}
        print("\nAluno adicionado com sucesso!")
    
# Função mostrar todos os alunos e suas respectivas notas no diário
def mostrar_diario(alunos):
    print("\nAlunos:")
    for ra, aluno in alunos.items():
        aluno = alunos[ra]
        nome = aluno['nome']
        notas = aluno['notas']
        print(f"{ra} {nome} | Notas: {notas}")

# Função pesquisar aluno pelo RA
def pesquisar_aluno(alunos, ra):
    if ra in alunos:
        aluno = alunos[ra]
        nome = aluno['nome']
        notas = aluno['notas']
        print(f"\nAluno encontrado!\n{ra} {nome} | Notas: {notas}")
    else:
        print("\nAluno não encontrado. Atenção ao digitar.")

# Função calcular média
def calcular_media(alunos, ra):
    if ra in alunos:
        aluno = alunos[ra]
        nome = aluno['nome']
        notas = aluno['notas']
        if len(notas) == 0:
            print(f"\nO aluno {nome} não possui notas registradas.")
        else:
            media = sum(notas) / len(notas)
            print(f"\n {ra} {nome} | Média - {media:.2f}")
    else:
        print("\nAluno não encontrado. Atenção ao digitar.")

# Função excluir aluno do diário
def excluir(alunos, ra):
    if ra in alunos:
        del alunos[ra]
        print("\nAluno excluido com sucesso")
    else:
        print("\nAluno não encontrado. Atenção ao digitar.")

# Menu do sistema
while True:
    print("\nDiário Professor - Gerenciamento de Alunos")
    print("\n1. Adicionar Aluno")
    print("2. Abrir Diário")
    print("3. Pesquisar Aluno por RA")
    print("4. Calcular Média Aluno")
    print("5. Excluir Aluno")
    print("6. Sair")

    escolha = input("Escolha uma opção: ")

    if escolha == '1':
        ra = input("Digite o RA do aluno: ")
        nome = input("Digite o nome do aluno: ")
        notas = [float(nota) for nota in input("Digite as notas do aluno separadas por espaço: ").split() if nota.replace('.', '', 1).isdigit()]
        novo_aluno(alunos, ra, nome, notas)
    elif escolha == '2':
        mostrar_diario(alunos)
    elif escolha == '3':
        ra = input("Digite o RA do aluno: ")
        pesquisar_aluno(alunos, ra)
    elif escolha == '4':
        ra = input("Digite o RA do aluno que deseja calcular a média das notas: ")
        calcular_media(alunos, ra)
    elif escolha == '5':
        ra = input("Digite o RA do aluno que deseja excluir: ")
        excluir(alunos, ra)
    elif escolha == '6':
        print("\nSaindo...")
        break
    else:
        print("\nOpção inválida. Por favor, escolha uma opção válida.")
```
